```python
# Put treatments into a list for for loop (and later for plot labels)

datafram = {'Capomulin': result_df_Capomulin['Tumor Volume (mm3)'],
            'Ramicane': result_df_Ramicane['Tumor Volume (mm3)'],
            'Infubinol':result_df_Infubinol['Tumor Volume (mm3)'],
            'Ceftamin':result_df_Ceftamin['Tumor Volume (mm3)']}

# Create empty list to fill with tumor vol data (for plotting)


# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for drug, current_dataframe in datafram.items():
    # Check if the column contains numeric values
    if pd.to_numeric(current_dataframe, errors='coerce').notna().all():
        # Calculate quartiles for the 'Tumor Volume (mm3)' column
        quartiles = current_dataframe.quantile([.25, .5, .75])

        # Calculate Interquartile Range (IQR)
        lowerq= quartiles[0.25]
        upperq= quartiles[0.75]
        iqr = quartiles[0.75] - quartiles[0.25]
        lower_bound = lowerq - (1.5*iqr)
        upper_bound = upperq + (1.5*iqr)
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
            
    # add subset 

    
    # Determine outliers using upper and lower bounds
        values_below_lowerq = current_dataframe.loc[current_dataframe < lower_bound]
        values_above_upperq = current_dataframe.loc[current_dataframe > upper_bound]
   
        # Print the results
        print(f"IQR for {drug}: {iqr}\n")
        print(f"Potential outliers: for {drug}\n{values_below_lowerq}\n and \n{values_above_upperq}\n")
```

## Correlation and Regression

```python
from scipy.stats import pearsonr
import statsmodels.api as sm



# Calculate the correlation coefficient
correlation_coefficient, _ = pearsonr(mouse_avg_tumor_reset['Weight (g)'], mouse_avg_tumor_reset['Tumor Volume (mm3)'])
print(f'Correlation Coefficient: {correlation_coefficient}')

# Fit a linear regression model
X = sm.add_constant(mouse_avg_tumor_reset['Weight (g)'])
y = mouse_avg_tumor_reset['Tumor Volume (mm3)']

model = sm.OLS(y, X).fit()

plt.scatter(mouse_avg_tumor_reset['Weight (g)'], mouse_avg_tumor_reset['Tumor Volume (mm3)'], label='Scatter Plot')
plt.plot(mouse_avg_tumor_reset['Weight (g)'], model.predict(X), color='red', label='Line of Best Fit')

plt.title('Weight vs Tumor Volume')
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.legend()
plt.show()
```
